### _Finding Outliers_

1. _IQR method_

```python
featured_data.plot(kind="box",subplots=True,layout=(7,2),figsize=(15,20))
plt.show()
```

```python
#OUTLIERS IDENTIFICATION USING IQR 
def iqr_outlier(x,factor):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    min_ = q1 - factor * iqr
    max_ = q3 + factor * iqr
    result_ = pd.Series([0] * len(x))
    result_[((x < min_) & (x > max_))] = 1
    return result_
```

```python
# plotting outliers
fig, ax = plt.subplots(7, 2, figsize=(20, 30))
row = col = 0
for n,i in enumerate(featured_data.columns):
    if (n % 2 == 0) & (n > 0):
        row += 1
        col = 0
    outliers = iqr_outlier(featured_data[i], 1.5)
   
    if sum(outliers) == 0:
        sns.scatterplot(x = np.arange(len(featured_data[i])), y = featured_data[i], ax = ax[row, col], legend=False, color = 'green')
    else:
        sns.scatterplot(x = np.arange(len(featured_data[i])), y = featured_data[i], ax = ax[row, col], hue = outliers, palette = ['green','red'])
    for x,y in zip(np.arange(len(featured_data[i]))[outliers == 1], featured_data[i][outliers == 1]):
        ax[row,col].text(x = x, y = y, s = y, fontsize = 8)
    ax[row,col].set_ylabel("")
    ax[row,col].set_title(i)
    ax[row,col].xaxis.set_visible(False)
    if sum(outliers) > 0:
        ax[row,col].legend(ncol=2)
    col += 1
ax[row,col].axis('off')
plt.show()
```

2. _Z Score method_

```python
def zscore_outlier(x,lb,ub):
    zscore = ((x - x.mean()) / x.std()).copy()
    result_ = pd.Series([0] * len(x))
    result_[((zscore < lb) & (zscore > ub))] = 1
    return result_
```

```python
#plotting outlier using z_score method
fig, ax = plt.subplots(7, 2, figsize=(20, 30))
row = col = 0
for n,i in enumerate(featured_data.columns):
    if (n % 2 == 0) & (n > 0):
        row += 1
        col = 0
    outliers = zscore_outlier(featured_data[i], -3, 3)
   
    if sum(outliers) == 0:
        sns.scatterplot(x = np.arange(len(featured_data[i])), y = featured_data[i], ax = ax[row, col], legend=False, color = 'green')
    else:
        sns.scatterplot(x = np.arange(len(featured_data[i])), y = featured_data[i], ax = ax[row, col], hue = outliers, palette = ['green','red'])
    for x,y in zip(np.arange(len(featured_data[i]))[outliers == 1], featured_data[i][outliers == 1]):
        ax[row,col].text(x = x, y = y, s = y, fontsize = 8)
    ax[row,col].set_ylabel("")
    ax[row,col].set_title(i)
    ax[row,col].xaxis.set_visible(False)
    if sum(outliers) > 0:
        ax[row,col].legend(ncol=2)
    col += 1
ax[row,col].axis('off')
plt.show()
```

### _Building Pipe lines for models_

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

1. _Pipeline using minmax scaler and SVC_

```python
pipe_svc_minmax = Pipeline([
    ('scaler', MinMaxScaler()),
    ('svc', SVC())
])
```

2. _Pipeline using standard scaler and SVC_

```python
pipe_svc_std = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC())
])
```

3. _Pipeline logistics regression_

```python
pipe_logisticR = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression())
])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier())
])
```

```python
from sklearn.linear_model import SGDClassifier

pipe_SGD = Pipeline([
    ('scaler', StandardScaler()),
    ('SGD', SGDClassifier(loss='hinge', penalty='l2', max_iter=1000, tol=1e-3))
])
```

```python
from sklearn.naive_bayes import GaussianNB

pipe_gnb = Pipeline([
    ('scaler', StandardScaler()),
    ('gnb', GaussianNB())
])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
pipes = [pipe_svc_minmax, pipe_svc_std, pipe_logisticR,pipe_SGD, pipe_rf, pipe_gnb]

for pipe in pipes:
    pipe.fit(X_train, y_train)
    print(pipe)
    print(f"Predict: {pipe.predict(X_test)}")
    print(f"Acc: {accuracy_score(pipe.predict(X_test), y_test)}")
    print(f"ROC AUC Score: {roc_auc_score(pipe.predict(X_test), y_test)}")
    print("===============================================")
```
